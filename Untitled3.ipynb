# Mais alguns testes

Z-teste para avaliar a diferença estatística entre os alunos com média acima e abaixo de 5.0.
O valor de 5.0 foi escolhido arbitrariamente e pode ser alterado para a nota de interesse. Por exemplo, caso estejamos interessados em prever alunos com chances de serem aprovados em cursos mais concorridos podemos usar G3 > 7.0

```
df_5plus = df[df['G3'] >= 5]
df_5minus = df[df['G3'] < 5]



t_stat, p_value = ztest(df_5plus.age, df_5minus.age)
print('Age\n')
print('p-value: ', p_value)
t_stat, p_value = ztest(df_5plus.sex, df_5minus.sex)
print('\nSex\n')
print('p-value: ', p_value)
t_stat, p_value = ztest(df_5plus.higher, df_5minus.higher)
print('\nHigher\n')
print('p-value: ', p_value)
t_stat, p_value = ztest(df_5plus.goout, df_5minus.goout)
print('\ngoout\n')
print('p-value: ', p_value)
```

Segundo os testes há uma diferença estatística entre as idades, a intenção de prosseguir com os estudos e o hábito de sair com os amigos dos alunos com boas notas(>5) e os com más notas.
O mesmo não foi confirmado para a variável de sexo.

```
matrix_df = pps.matrix(df)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
matrix_df = matrix_df.apply(lambda x: round(x, 2)) # Rounding matrix_df's values to 0,XX
plt.subplots(figsize = (15,15))

sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=0.75, annot=True)
```

A análise do PPS(predictive power score) acima permite ter uma ideia do poder preditivo de certas variáveis mesmo quando a relação não é linear, ao contrario da matriz de correlações que necessita de relação linear).
